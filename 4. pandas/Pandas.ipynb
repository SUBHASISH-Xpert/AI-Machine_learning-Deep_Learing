# Merging , joining and concatination

```python
import numpy as np
import pandas as pd
```

```python
employees = pd.DataFrame({
    "employe_id":[1,2,3,4,5],
    "name":['john','rahul','lina','bob','peter'],
    "department":['HR','IT','FINANCE','IT','HR']
})

salaries = pd.DataFrame({
    'employe_id':[1,2,3,6,7],
    'salary':[60000,49000,66000,56000,90000],
    'bonus':[5000,10000,7000,5000,12000]
})
```

```python
employees
```

```python
salaries
```

```python
# common 1,2,3 i see that
# mearging  ->only common give 
pd.merge(employees,salaries)
```

```python
# how = default -> inner, 'outer' -> all data missing value nan
pd.merge(employees,salaries, on ='employe_id', how = 'inner')
```

```python
# all are but not that nan value
pd.merge(employees,salaries, on ='employe_id', how = 'outer')
```

```python
# employe_id  left all id shown in 1st data frame 
pd.merge(employees,salaries, on ='employe_id', how = 'left')
```

```python
# right means second data all are present with common and fill nan
pd.merge(employees,salaries, on ='employe_id', how = 'right')
```

# Groupby and Aggregation

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'categort':['A','B','A','B','A','B','A','B'],
    'store': ['51','51','52','52','51','52','52','51'],
    'sales':[100,200,150,250,120,180,200,300],
    'Quantity':[10,15,12,18,8,20,15,25],
    'data':pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
df
```

```python
# group by category and calculate the sum of sales using gropby
cat = df.groupby('categort')
for i, v in cat :
    print(i)
    print(v)
```

```python
cat = df.groupby('categort')['sales'].sum()
cat
```

```python
# groupby store and calculate the sum of sales 
cat1 = df.groupby('store')['sales'].sum()
cat1
```

```python
# group by multiple columns
# group by category and store
cat2 = df.groupby(['categort','store'])['sales'].sum()
cat2
```
