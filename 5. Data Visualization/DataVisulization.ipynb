# Regression Plot

# lmplot() -> regression plot is use in machine learning so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we can put inside it

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.lmplot(x = 'total_bill', y ='tip',  data = tips)
```

```python
# mail and female line
sns.lmplot(x ='total_bill', y ='tip', data = tips, hue = 'sex')
plt.show()
```

```python
sns.lmplot(x ='total_bill', y ='tip', data = tips, hue = 'sex',palette = 'coolwarm')
plt.show()
```

```python
sns.lmplot(x ='total_bill', y ='tip', data = tips, hue = 'sex', palette = 'coolwarm',markers = ['o','v'])
plt.show()
```

```python
sns.lmplot(x ='total_bill', y ='tip', data = tips, hue = 'sex', palette = 'coolwarm',markers = ['o','v'], scatter_kws = {'s':100})
plt.show()
```
